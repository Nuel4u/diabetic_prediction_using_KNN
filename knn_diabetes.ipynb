```python
zero_not_accepted = ['Glucose' ,'BloodPressure','SkinThickness','Insulin','BMI']

for column in zero_not_accepted:
    diabetes[column] =diabetes[column].replace(0 , np.NaN)
    mean = int(diabetes[column].mean(skipna=True))
    diabetes[column] = diabetes[column].replace(np.NaN ,mean)
```

```python
x = diabetes.iloc[:,0:8]   # : meaning all the rows. 0:8 from 0 to 7 
y = diabetes.iloc[:,8]    # : all rows and the 8th row
x_train,x_test ,y_train,y_test = train_test_split(x,y ,random_state =0 ,test_size =0.2)
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
y_test = sc.transform(x_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors =11 , p=2 ,metric ='euclidean')
```

```python
classifier.fit(x_train ,y_train)
```

```python
y_pred =classifier.predict(y_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test ,y_pred)
print(cm)
```

```python
print(f1_score(y_test , y_pred))
```

```python
accuracy_score(y_test,y_pred)
```
