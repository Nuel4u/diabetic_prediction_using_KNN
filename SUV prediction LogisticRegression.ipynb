# Training the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train,y_test = train_test_split(x,y ,test_size =0.25 , random_state =0)
```

```python
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()                                       # scaling over a large amoutn of data
```

```python
x_train = sc.fit_transform(x_train)      # scaling over the trained , fitted transformed data
x_test = sc.transform(x_test)           # scales through test data
```

```python
from  sklearn.linear_model import LogisticRegression
```

```python
classfier = LogisticRegression(random_state = 0)     
```

```python
classfier.fit(x_train ,y_train)        # fitting x_train and y_train 
```

```python
y_pred = classfier.predict(x_test)    # predicting tested data
```

```python
from sklearn.metrics import accuracy_score   # getting the accurate figure
```

```python
accuracy_score(y_test ,y_pred)            # getting the accurate figure
```

```python
accuracy_score(y_test ,y_pred)*100
```
